## a. исследование последовательностей на монотонность

$$1.\,\,\,\{a_n\}_{n=1}^\infty=2^n-n$$

$$a_1 = 2^1 - 1 = 1$$   
$$a_2 = 2^2 - 2 = 2$$  
$$a_3 = 2^3 - 3 = 5$$   
$$a_1 < a_2 < a_3$$ 

$$монотонное \ возрастание$$


$$2.\,\,\,\{b_n\}_{n=2}^\infty=\frac{1}{1 - n}$$ 

$$b_2 = \frac{1}{1 - 2} = -1$$   
$$b_3 = \frac{1}{1 - 3} = -\frac{1}{2}$$   
$$b_4 = \frac{1}{1 - 4} = -\frac{1}{3}$$   
$$b_2 < b_3 < b_4$$ 

$$монотонное \ возрастание$$

$$3.\,\,\,\{c_n\}_{n=1}^\infty=(-1)^n + \sqrt{2n}$$
 
$$c_1 = (-1)^{1} + \sqrt{2\cdot 1} = -1 + \sqrt{2} \approx 0,4$$  
$$c_2 = (-1)^2 + \sqrt{2\cdot 2} = 1 + 2 = 3$$   
$$c_3 = (-1)^3 + \sqrt{2\cdot 3} = -1 + \sqrt{6} \approx 1,45$$   
$$c_1 < c_2 > c_3$$

$$последовательность \ немонотонная$$

$$4.\,\,\,\{d_n\}_{n=1}^\infty=(-1)^{2n} + \frac{1}{n^2}$$  
 
$$d_1 = (-1)^{2} + \frac{1}{1^2} = 1 + 1 = 2$$  
$$d_2 = (-1)^{4} + \frac{1}{2^2} = 1 + 0,25 = 1,25$$  
$$d_3 = (-1)^{6} + \frac{1}{3^2} = 1 + \frac{1}{9} = 1\frac{1}{9}$$  
$$d_1 > d_2 > d_3$$

$$монотонное \ убывание$$

## b. исследование последовательностей на ограниченность;

$$1.\,\,\,\{a_n\}_{n=1}^\infty=2^n-n$$

$$n=1: \,\,\, a_1 = 2^1-1 = 1$$

Члены последовательности принимают значения на интервале [1; +∞), следовательно последовательность ограничена снизу.

$$2.\,\,\,\{b_n\}_{n=2}^\infty=\frac{1}{1 - n}$$ 

$$n=2: \,\,\, b_2 = \frac{1}{1 - 2} = -1$$  

Члены последовательности принимают значения на интервале (-∞; -1], следовательно последовательность ограничена сверху.

$$3.\,\,\,\{c_n\}_{n=1}^\infty=(-1)^n + \sqrt{2n}$$

$$n=1: \,\,\, c_1 = (-1)^{1} + \sqrt{2\cdot 1} = -1 + \sqrt{2}$$

Члены последовательности принимают значения на интервале [sqrt(2)-1; +∞), следовательно последовательность ограниченна снизу.

$$4.\,\,\,\{d_n\}_{n=1}^\infty=(-1)^{2n} + \frac{1}{n^2}$$ 

$$n=1: \,\,\, d_1 = (-1)^{2} + \frac{1}{1^2} = 1 + 1 = 2$$ 

$$n\to \infty:$$ $$\lim_{n\to \infty} \bigr((-1)^{2n} + \frac{1}{n^2}\bigl) = 1 + 0 = 1$$

Члены последовательности принимают значения на интервале [2; 1), следовательно последовательность ограниченна.
